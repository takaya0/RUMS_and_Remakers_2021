```python
import numpy as np
```

```python
def sym(A):
    return (A + A.T)/2
```

```python
while True:
    arr = np.random.randint(-10, 31, (10, 10))
    sym_arr = sym(arr)
    max_eign = np.max(np.linalg.eigh(sym_arr)[0])
    if max_eign == int(max_eign):
        print(sym_arr)
        print(max_eign)
        break
```
